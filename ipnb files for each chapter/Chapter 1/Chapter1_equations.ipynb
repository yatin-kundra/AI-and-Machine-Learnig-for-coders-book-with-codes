# This is a model that takes x and y and get trained


## importing dependencies

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
# print(tf.__version__)
```

```python
model = Sequential([Dense(units=1, input_shape=[1])])

# (*) sequential is t means that the layers in the model are arranged sequentially,
#   one after another, forming a linear flow of data through the network.

# (*) Dense means that each neuron in this layer is connected to every single neuron
#   in the previous layer.

# (*) unints = 1 means that there is only one neuron in input layer of the model ,
#   i.e. dense layer will have only one neurone.

# (*) input_shape = [1]  indicates that the input layer will have only one feature

model.compile(optimizer="sgd", loss = "mean_squared_error") # this lines means compiling the model
#specifying the optimizer and loss functin for the Model

## sgd(stochastic gradient descent ) is a popular optimizer for the neural networks
## mean_squared_error (MES = mean squared error) is common loss function for regressin models
```

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0] , dtype = float)  # independent variable
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype = float)  # dependent variable
```

```python
model.fit(xs, ys, epochs = 500)   # this line means training the model for 500 epoches
# xs and ys should be of type np array
# (*) epochs = 500 means , epoches tells how many times will the model iterate and train over the given dataset.
print("the ans is " ,model.predict([4.0]))
```
